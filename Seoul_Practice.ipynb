## Scraping the first page

```python
#첫번째 기사 링크 선택해서 XPath로 살펴보기(Copy > Copy Xpath)
#Ctrl+V
#//*[@id="content"]/div[2]/div/div[1]/div[1]/a : 첫번째 기사 링크 xpath
#//*[@id="content"]/div[2]/div/div[1]/div/a : 전체 기사 링크 xpath

a_path = '//*[@id="content"]/div[2]/div/div[1]/div/a'
```

```python
news = ch_driver.find_elements_by_xpath(a_path)
news[0].text
```

```python
#get link
post_link = news[0].get_attribute('href')
post_link
```

```python
#get subject, post id
subject = news[0].find_element_by_class_name('subject').text.split(']')
post_id = subject[0].strip('[')
post_id
subject = subject[1].strip()
print(post_id, subject)
```

```python
#get date, time, category
items = news[0].find_element_by_class_name("date").text.split()
post_time = items[0]+" "+items[1]
post_cat = items[2].strip("[]")
print(post_time, post_cat)
```

```python
for item in news:
    #get link
    post_link = item.get_attribute('href')
    
    #get subject, post id
    subject = item.find_element_by_class_name('subject').text.split(']')
    post_id = subject[0].strip('[')
    subject = subject[1].strip()
    
    #get date-time, category
    items = item.find_element_by_class_name("date").text.split()
    post_time = items[0]+" "+items[1]
    post_cat = items[2].strip("[]")
    
    #get content
    content = item.find_element_by_class_name("txt").text
    
    print(post_link, post_id, subject, post_time, post_cat, content, sep="\n")
    print("\n")
```

```python
#make a function of it

def get_article():
    news = ch_driver.find_elements_by_xpath(a_path)
    result = []
    
    for item in news:
        #get link
        post_link = item.get_attribute('href')
    
        #get subject, post id
        subject = item.find_element_by_class_name('subject').text.split(']')
        post_id = subject[0].strip('[')
        subject = subject[1].strip()
    
        #get date-time, category
        items = item.find_element_by_class_name("date").text.split()
        post_time = items[0]+" "+items[1]
        post_cat = items[2].strip("[]")
    
        #get content
        content = item.find_element_by_class_name("txt").text
    
        result.append((post_link, post_id, subject, post_time, post_cat, content))
        
    return result
```

## Scraping through to 10th page

```python
# ch_driver.execute_script('fnPagingMove(2);') => can move to page 2
# so, fnPagingMove(n) => 1 to 10

from time import sleep

# tell driver to wait for 5 secs when certain elements are not available immediately.
ch_driver.implicitly_wait(5)

all_result = []
for i in range(1, 10+1):
    #move to page i
    ch_driver.execute_script('fnPagingMove({page_number});'.format(page_number = i))
    
    #after 1sec sleep
    sleep(1)
    
    #get all contents and append them to all_result
    all_result.append(get_article())
    
all_result
```
